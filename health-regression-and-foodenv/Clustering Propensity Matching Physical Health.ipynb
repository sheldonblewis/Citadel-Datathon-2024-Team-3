```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import NearestNeighbors
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Load your data
data = pd.read_csv("merged_regression_data.csv")

# Define the features for clustering and the features of interest
clustering_features = [
    'Rural_2011', 
    'Household Income_2011', 
    'African American_2011', 
    'Asian_2011', 
    'Hispanic_2011'
    
]

features_of_interest = ['Percent_change_fast_food_restaurants_2011_2016', 'change_fair_poor_health']

# Drop rows with missing values in the clustering features or features of interest
data = data.dropna(subset=clustering_features + features_of_interest)

# Define treatment (significant increase in fast food restaurants)
data['treatment'] = (data['Percent_change_fast_food_restaurants_2011_2016'] > 3).astype(int)

# Standardize the clustering features
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data[clustering_features])

# Perform KMeans clustering
kmeans = KMeans(n_clusters=5, random_state=42)  # Adjust the number of clusters as needed
data['cluster'] = kmeans.fit_predict(data_scaled)

# Function to perform PSM and t-test for each cluster
def psm_and_ttest_cluster(cluster_data, cluster_id):
    # Select only numeric columns for the Logistic Regression model
    numeric_cols = cluster_data.select_dtypes(include=['float64', 'int64']).columns
    columns_of_interest = numeric_cols.drop(['treatment', 'change_fair_poor_health'])
    
    # Drop rows with NaN values in the columns of interest
    cluster_data = cluster_data.dropna(subset=columns_of_interest)
    
    X = cluster_data[columns_of_interest]
    y = cluster_data['treatment']
    
    # Fit Logistic Regression model
    log_reg = LogisticRegression()
    log_reg.fit(X, y)
    
    # Calculate propensity scores
    cluster_data['propensity_score'] = log_reg.predict_proba(X)[:, 1]
    
    # Nearest Neighbors Matching
    nn = NearestNeighbors(n_neighbors=1)
    treated = cluster_data[cluster_data['treatment'] == 1]
    control = cluster_data[cluster_data['treatment'] == 0]
    
    nn.fit(control[['propensity_score']])
    distances, indices = nn.kneighbors(treated[['propensity_score']])
    
    matched_control = control.iloc[indices.flatten()]
    matched_data = pd.concat([treated, matched_control])
    
    # Perform t-test
    treated_obese = matched_data[matched_data['treatment'] == 1]['change_fair_poor_health']
    control_obese = matched_data[matched_data['treatment'] == 0]['change_fair_poor_health']
    
    t_stat, p_value, _ = sm.stats.ttest_ind(treated_obese, control_obese)
    
    # Plot results
    plt.figure(figsize=(8, 6))
    sns.boxplot(x='treatment', y='change_fair_poor_health', data=matched_data, palette="Set3")
    plt.title(f"Cluster: {cluster_id} - T-test: t={t_stat:.2f}, p={p_value:.3f}")
    plt.xlabel('Treatment (1 = Increased Fast Food Restaurants)')
    plt.ylabel('Change in Poor Health')
    plt.xticks([0, 1], ['Control', 'Treated'])
    plt.tight_layout()
    plt.savefig(f"cluster_{cluster_id}_fast_food_mental_health.png")
    plt.show()
    
    return t_stat, p_value

# Loop over each cluster and perform the analysis
results = {}
for cluster_id in data['cluster'].unique():
    cluster_data = data[data['cluster'] == cluster_id]
    if len(cluster_data['treatment'].unique()) > 1:  # Ensure there are both control and treated groups
        print(f"Analyzing Cluster: {cluster_id}")
        t_stat, p_value = psm_and_ttest_cluster(cluster_data, cluster_id)
        results[cluster_id] = {'t_stat': t_stat, 'p_value': p_value}

# Visualizing the P-values across clusters
clusters = list(results.keys())
p_values = [results[cluster]['p_value'] for cluster in clusters]

plt.figure(figsize=(12, 8))
sns.barplot(x=clusters, y=p_values, palette='viridis')
plt.axhline(y=0.05, color='red', linestyle='--')
plt.title('P-values of T-test for Change in Obesity by Cluster Upon Introduction of Fast Food')
plt.xlabel('Cluster')
plt.ylabel('P-value')
plt.xticks(rotation=90)
plt.tight_layout()
plt.savefig("propensity_matching_clusters_poorhealth.png")
plt.show()

# Display the results
results_df = pd.DataFrame(results).T
print(results_df)
```
