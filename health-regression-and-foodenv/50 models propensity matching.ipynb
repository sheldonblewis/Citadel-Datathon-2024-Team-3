```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import NearestNeighbors

# Load your data
data = pd.read_csv("merged_regression_data.csv")

# Drop rows with missing values in the columns we're interested in
data = data.dropna(subset=['Percent_change_fast_food_restaurants_2011_2016', 'change_mentally_unhealthy_days'])

# Define treatment (significant increase in fast food restaurants)
data['treatment'] = (data['Percent_change_fast_food_restaurants_2011_2016'] > 3).astype(int)

# Define control group (no significant increase in fast food restaurants)
control_group = data[data['treatment'] == 0]

# Define treated group (significant increase in fast food restaurants)
treated_group = data[data['treatment'] == 1]
```

```python

# Function to perform PSM and t-test for each state
def psm_and_ttest(state_data, state_name):
    # Select only numeric columns for the Logistic Regression model
    numeric_cols = state_data.select_dtypes(include=['float64', 'int64']).columns
    columns_of_interest = numeric_cols.drop(['treatment', 'change_mentally_unhealthy_days'])
    
    # Drop rows with NaN values in the columns of interest
    state_data = state_data.dropna(subset=columns_of_interest)
    
    X = state_data[columns_of_interest]
    y = state_data['treatment']
    
    # Fit Logistic Regression model
    log_reg = LogisticRegression()
    log_reg.fit(X, y)
    
    # Calculate propensity scores
    state_data['propensity_score'] = log_reg.predict_proba(X)[:, 1]
    
    # Nearest Neighbors Matching
    nn = NearestNeighbors(n_neighbors=1)
    treated = state_data[state_data['treatment'] == 1]
    control = state_data[state_data['treatment'] == 0]
    
    nn.fit(control[['propensity_score']])
    distances, indices = nn.kneighbors(treated[['propensity_score']])
    
    matched_control = control.iloc[indices.flatten()]
    matched_data = pd.concat([treated, matched_control])
    
    # Perform t-test
    treated_obese = matched_data[matched_data['treatment'] == 1]['change_mentally_unhealthy_days']
    control_obese = matched_data[matched_data['treatment'] == 0]['change_mentally_unhealthy_days']
    
    t_stat, p_value, _ = sm.stats.ttest_ind(treated_obese, control_obese)
    
    # Plot results
    plt.figure(figsize=(8, 6))
    sns.boxplot(x='treatment', y='change_mentally_unhealthy_days', data=matched_data, palette="Set3")
    plt.title(f"State: {state_name} - T-test: t={t_stat:.2f}, p={p_value:.3f}")
    plt.xlabel('Treatment (1 = Increased Fast Food Restaurants)')
    plt.ylabel('Change in Obesity')
    plt.xticks([0, 1], ['Control', 'Treated'])
    plt.tight_layout()
    plt.show()
    
    return t_stat, p_value

# Loop over each state and perform the analysis
results = {}
for state in data['State_2011'].unique():
    state_data = data[data['State_2011'] == state]
    if len(state_data['treatment'].unique()) > 1:  # Ensure there are both control and treated groups
        print(f"Analyzing State: {state}")
        t_stat, p_value = psm_and_ttest(state_data, state)
        results[state] = {'t_stat': t_stat, 'p_value': p_value}
```

```python
# Visualizing the P-values across states
states = list(results.keys())
p_values = [results[state]['p_value'] for state in states]

plt.figure(figsize=(12, 8))
sns.barplot(x=states, y=p_values, palette='viridis')
plt.axhline(y=0.05, color='red', linestyle='--')
plt.title('P-values of T-test for Change in Mental Health by State Upon Introduction of Fast Food')
plt.xlabel('State')
plt.ylabel('P-value')
plt.xticks(rotation=90)
plt.tight_layout()
plt.savefig("propensity matching introduction of fast food mental health")
plt.show()

# Display the results
results_df = pd.DataFrame(results).T
print(results_df)
```
